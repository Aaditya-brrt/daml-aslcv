```python
import tensorflow as tf
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "archive/asl_alphabet_train/asl_alphabet_train",
    labels='inferred',
    label_mode='int',
    class_names=None,
    color_mode='rgb',
    batch_size=32,
    image_size=(256, 256),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False,
    pad_to_aspect_ratio=False,
    data_format=None,
    verbose=True
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "archive/asl_alphabet_test/asl_alphabet_test",
    labels='inferred',
    label_mode='int',
    class_names=None,
    color_mode='rgb',
    batch_size=32,
    image_size=(256, 256),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False,
    pad_to_aspect_ratio=False,
    data_format=None,
    verbose=True
)
```

```python
print(train_ds.class_names)
print(test_ds.class_names)
```

```python
val_batches = int(0.5 * tf.data.experimental.cardinality(test_ds))

val_ds = test_ds.take(val_batches)
test_ds = test_ds.skip(val_batches)
```

```python
#process data

for images, labels in train_ds.take(1):
    print(images.shape)  # (batch_size, height, width, channels)
    print(labels.shape)  # (batch_size,)

```

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)

train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
test_ds = test_ds.map(lambda x, y: (normalization_layer(x), y))
val_ds = val_ds.map(lambda x, y: (normalization_layer(x), y))
```
